```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from random import *
from matplotlib import pyplot as plt
import numpy as np
import GPy
from datetime import datetime

m_train_set_path = "./MALLAT/MALLAT_TRAIN"
```

```python
# class name {1,2,3}
class_names = np.loadtxt(m_train_set_path, delimiter=',', usecols=[0])

m_data = np.loadtxt(m_train_set_path, delimiter=',',usecols=range(1,1025))

# normalize magnitude
m_data -= m_data.mean(1)[:,np.newaxis]
m_data /= m_data.std(1)[:,np.newaxis]
```

```python
plt.figure(figsize=(10,2))
X_r = np.arange(1,1025)
unique_class_names = np.unique(class_names)

for un_class_index, class_n in enumerate([3,6,7]):
    ax = plt.subplot(1,3,un_class_index+1)
    for indx, c in enumerate(class_names):
        if(c == class_n):
            print(indx)
            ax.plot(X_r, m_data[indx,:])
            plt.xlabel('class %i'%class_n)
            break;
            
            
plt.subplots_adjust(bottom=0.3, top=0.95)
fig_name = './MALLAT_test_result/demo.png';
plt.savefig(fig_name, format='png', dpi=500);
```
